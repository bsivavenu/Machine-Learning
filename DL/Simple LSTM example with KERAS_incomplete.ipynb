```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
import numpy as np
```

```python
data = [i for i in range(100)]
target = [i for i in range(1,101)]
data = np.array(data,dtype=float)
target = np.array(target,dtype=float)
```

```python
data = data.reshape((1,1,100))
target = target.reshape((1,1,100))
```

```python
x_test = [i for i in range(200,300)]
y_test = [i for i in range(201,301)]

x_test = np.array(x_test).reshape(1,1,100)
y_test = np.array(y_test).reshape(1,1,100)
```

```python
model =Sequential()
model.add(LSTM(100,input_shape=(1,100),return_sequences=True))
model.add(Dense(100,activation='relu'))
# model.add
model.compile(loss='mean_absolute_error',optimizer='adam',metrics=['accuracy'])
model.fit(data,target,epochs=500,batch_size=1,validation_data=(x_test,y_test))
```

```python
model.predict(data)
```

```python
score = model.evaluate(data,target)
score[0]
```

In the above example we haven't used normalization and activation function and type of loss. Now we'll change the params to get good accuracy.

```python
data = data.reshape((1,1,100))/200
target = target.reshape((1,1,100))/200
x_test = [i for i in range(200,300)]
y_test = [i for i in range(201,301)]

x_test = np.array(x_test).reshape(1,1,100)
y_test = np.array(y_test).reshape(1,1,100)
```

```python
model =Sequential()
model.add(LSTM(100,input_shape=(1,100),return_sequences=True,activation='relu'))
# model.add(Dense(100,activation='relu'))
# model.add
model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
model.fit(data,target,epochs=500,batch_size=1,validation_data=(data,target))
```

```python
score = model.evaluate(data,target)
score[0]
```
