```python
from nltk.corpus import stopwords
```

```python
def clean_doc(doc):
    tokens = doc.split()
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not  w in stop_words]
    tokens = [w for w in tokens if len(w)>1]
    tokens = [w for w in tokens if w.isalpha()]
    return tokens
```

```python
from os import listdir
from collections import Counter
```

```python
def add_doc_to_vocab(filename,vocab):
    text = load_file(filename)
    tokens = clean_doc(text)
    vocab.update(tokens)
```

```python
def process_docs(directory,vocab):
    for filename in listdir(directory):
        if filename.startswith('cv9'):
            continue
        path = directory + '/' + filename

        add_doc_to_vocab(path,vocab)
```

```python
def load_doc(filename):
    file = open(filename,'r')
    text = file.read()
    file.close()
    return text
vocab_file = 'c:/Users/HP/Desktop/vocab.txt'
vocab = load_doc(vocab_file)
vocab = vocab.split()
vocab =set(vocab)
print(vocab)
```

```python
def doc_to_line(filename, vocab):
	# load the doc
	doc = load_doc(filename)
	# clean doc
	tokens = clean_doc(doc)
	# filter by vocab
	tokens = [w for w in tokens if w in vocab]
	return ' '.join(tokens)
# load all docs in a directory
def process_docs(directory, vocab):
	lines = list()
	# walk through all files in the folder
	for filename in listdir(directory):
		# skip files that do not have the right extension
		if not filename.endswith(".txt"):
			continue
		# create the full path of the file to open
		path = directory + '/' + filename
		# load and clean the doc
		line = doc_to_line(path, vocab)
		# add to list
		lines.append(line)
	return lines
# prepare negative reviews
negative_lines = process_docs('c:/Users/HP/Desktop/movie_reviews/neg', vocab)
save_list(negative_lines, 'negative.txt')
# prepare positive reviews
positive_lines = process_docs('c:/Users/HP/Desktop/movie_reviews/pos', vocab)
save_list(positive_lines, 'positive.txt')
```
