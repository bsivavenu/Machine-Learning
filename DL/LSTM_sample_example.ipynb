```python
from keras.layers import Dense,LSTM
from  keras.models import Sequential,load_model
from numpy import array
import numpy as np
```

```python
def get_train():
    seq = [[0.0,0.1],[0.1,0.2],[0.2,0.3],[0.3,0.4],[0.4,0.5],[0.5,0.6]]
    seq = array(seq)
    X,y = seq[:,0],seq[:,1]
    X = X.reshape(len(X),1,1)
    return X,y
```

```python
model = Sequential()
model.add(LSTM(12,input_shape=(1,1)))
model.add(Dense(1,activation='linear'))
```

```python
model.compile('adam',loss='mse')
# model.fit()
```

```python
X,y = get_train()
```

```python
X.shape,y.shape
```

```python
model.fit(X,y,verbose=1,epochs=300,shuffle=False)
```

```python
model.save('lstm_sample.h5')
```

```python
model.load_weights('lstm_sample.h5')
```
